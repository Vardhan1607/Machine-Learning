# ***`Task1.1 - Vocabulary without Stop-words`***

```
#Finding the stop words
nltk.download('stopwords')
stop_words = stopwords.words('english')

#Cateogrizing our Data
science_txt,sports_txt,business_txt,covid_txt='','','',''
for i in range(data.shape[0]):
  if(data['category'][i] =='science'):
    science_txt+= data['text'][i] +' '
  elif(data['category'][i]=='sports'):
    sports_txt+=data['text'][i]+' '
  elif(data['category'][i]=='business'):
    business_txt+=data['text'][i]+' '
  elif(data['category'][i]=='covid'):
    covid_txt+=data['text'][i]+' '
```

```
#Removing Digits and some special Characters
def remove_digits(my_text):
  try:
    my_text = my_text.replace('”',"")
    my_text = my_text.replace('’',"")
    my_text = my_text.replace('“',"")
    my_text = my_text.replace('—',"")
  except:
    pass
  res = ''.join([i for i in my_text if not i.isdigit()])
  return res
```

```
#Removing digits from our texts
science_txt = remove_digits(science_txt)
sports_txt = remove_digits(sports_txt)
business_txt = remove_digits(business_txt)
covid_txt = remove_digits(covid_txt)
```

```
#Function to remove stop words
def stopwords(text):
  words = text.split()
  #Removing punctuations
  table = str.maketrans('', '', string.punctuation)
  stripped = [w.translate(table) for w in words]
  words = stripped
  words = [word.lower() for word in words]
  #Removing the stop words
  filtered_sentence = []   
  for w in words: 
    if w not in stop_words and w!='':
        filtered_sentence.append(w)
  #Returning the list after removing stop words
  return filtered_sentence
#Printing the stop words
print('The Stop Words ')
print(stop_words) 
```

```
#Printing science words after removing stop words
science_filter = stopwords(science_txt)
print("Vocabulary in Science without Stop Words")
print(science_filter)
```

```
#Printing sports words after removing stop words
sports_filter = stopwords(sports_txt)
print("Vocabulary in Sports without Stop Word")
print(sports_filter)
```

```
#Printing business words after removing stop words
business_filter = stopwords(business_txt)
print("Vocabulary in Business without Stop Word")
print(business_filter)
```

```
#Printing covid words after removing stop words
covid_filter = stopwords(covid_txt)
print("Vocabulary in COVID without Stop Word")
print(covid_filter)
```

```
#Printing all words after removing stop words
final_txt = science_txt + covid_txt + business_txt + sports_txt
final_filter = stopwords(final_txt)
print("Final Vocabulary without Stop Word")
print(final_filter)
```

```
#Function to return the Vocabulary
def vocab(filter):
  vectorizer = CountVectorizer()
  vectorizer.fit(filter)
  print(vectorizer.vocabulary_)
  vector = vectorizer.transform(stop_words)
  return vector
```

```
#Printing the Science Vocabulary
print('Vocabulary in science')
vocab(science_filter)
#Printing the Sports Vocabulary
print('')
print('Vocabulary in sports')
vocab(sports_filter)
#Printing the Business Vocabulary
print('')
print('Vocabulary in business')
vocab(business_filter)
#Printing the Covid Vocabulary
print('')
print('Vocabulary in covid')
vocab(covid_filter)
#Printing the Final Vocabulary
print('')
print('Final Vocabulary')
vocab(final_filter)
```

# ***`Task 1.3 - Class-Conditional Probabilities`***

```
#Function to calculate the frequency
def frequency_of_words(filtered):  
  frequency = {} 
  for items in set(filtered): 
    frequency[items] = filtered.count(items) 
  return frequency
```

```
final_cc=frequency_of_words(final_filter)

#Function for finding the class conditional probability
def class_conditional_probability(class_frequency):
  class_freq=frequency_of_words(class_frequency)
  class_cc = {}
  n=len(class_frequency)
  for word,freq in final_cc.items():
    if(word not in class_freq):
      #If word not found in that label , assigning the probability equal to 1e-17
      #Else the CCP wont be equal to 1
      class_cc[word]=1e-17
    else:
      #If word found ,the frequency of word/total words in class
      class_cc[word]=(class_freq[word]/n)
  return class_cc
```

```
#Class Conditional Probability of Label: science
science_cc = class_conditional_probability(science_filter)
print('Class conditional Probability of each word in Science')
print(science_cc)

#Class Conditional Probability of Label: science
sports_cc = class_conditional_probability(sports_filter)
print('Class conditional Probability of each word in Sports')
print(sports_cc)

#Class Conditional Probability of Label: business
business_cc = class_conditional_probability(business_filter)
print('Class conditional Probability of each word in Business')
print(business_cc)

#Class Conditional Probability of Label: covid
covid_cc = class_conditional_probability(covid_filter)
print('Class conditional Probability of each word in Covid')
print(covid_cc)
```

```
#Representing the CCP in a DataFrame
final_cdp={}
for word,freq in science_cc.items():
  lst=[freq,sports_cc[word],business_cc[word],covid_cc[word]]
  final_cdp[word]=lst
df = pd.DataFrame(final_cdp, index =['Science', 'Sports', 'Business', 'Covid'])
df
```

# ***`Task 1.4 -Posterior Distribution over the labels`***

```
#Reading the Test Data
test_data =pd.read_csv('testdata.csv')
test_data
```

```
#Function to Find the label of the sentence
def label_divide(my_sentence):
  my_filtered = stopwords(remove_digits(my_sentence))
  science_probab,business_probab,sports_probab,covid_probab=1,1,1,1
  #Looping for each word
  for word in my_filtered:
    if(word in science_cc):
      science_probab*= science_cc[word]
    if(word in sports_cc):
      sports_probab*= sports_cc[word]
    if(word in business_cc):
      business_probab*= business_cc[word]
    if(word in covid_cc):
      covid_probab*= covid_cc[word]
  #Finding the max probability
  my_label = max(science_probab,sports_probab,business_probab,covid_probab)
  #Finding the k-value
  k_value = 1/(science_probab+sports_probab+business_probab+covid_probab)
  #Returning accordingly
  if(my_label ==science_probab):
    return 'science',(k_value)*science_probab
  elif(my_label==sports_probab):
    return 'sports',(k_value)*sports_probab
  elif(my_label == business_probab):
    return 'business',(k_value)*business_probab
  elif(my_label==covid_probab):
    return 'covid',(k_value)*covid_probab
```

```
#Finding the Posterior distribution of the test data
def posterior_distribution(my_data):
  post_distribute=[]
  for i in range(my_data.shape[0]):
    x=label_divide(my_data['text'][i])
    post_distribute.append([x[0],x[1]])
  final_frame = pd.DataFrame(post_distribute,columns=['Predicted Category','Posterior Distribution'])
  return post_distribute,pd.concat([test_data,final_frame],axis=1)
```

```
list_pd,data_pd = posterior_distribution(test_data)
list_pd
```

# ***`Task 2.2 - Prior Probabilities of all "labels" `***

```
#Function to find the frequency
def frequency(word,list1):
  a=0
  for element in list1:
    if (element==word):
      a=a+1
  return a

#Calculating the Prior Probability
abc = list(dict.fromkeys(full_vocab))
prior={}
for word in abc:
  prior[word]=frequency(word,full_vocab)/len(full_vocab)
print(prior)  
```

# ***`Task 2.3 - Class-Conditional Probabilities of all words`***

```
#Function foe finding conditional probability
def conditional_prob(a,b,arr):
  k=0
  for list1 in arr:
    if((a in list1) and (b in list1)):
      k=k+1
  if(k==0):
    return 0.0000000000001
  return k

def frqqq(d,arr):
  f=0
  for list1 in arr:
    if(d in list1):
      f=f+1
  return f                        
 
```

```
#Finding the class conditional probability
cond={}
for lab in abc:
  cond[lab]={}
  for word in abc:
    if(word==lab):
      continue
    cond[lab][word]=conditional_prob(word,lab,tr_vocab)/frqqq(lab,tr_vocab)
print(cond)  
```

```
#Finding the posterior distribution
def go(w,cp,b,c):
  a=1
  a=a*b[w]
  for item in c:
    #If we have found the word , then we multiply the word
    try:
      a=a*cp[w][item]
    #Else multiply with a smaller value
    except:
      a=a*0.00000000000000001
  if (a==1):
    return 0.000000000000000001
  return a
#Calculating the posterior distribution
find=[]
i=0
for sentence in bl_vocab:
  find.append({})
  for word in abc:
    find[i][word]=go(word,cond,prior,sentence)
  i=i+1
print(find)
```
